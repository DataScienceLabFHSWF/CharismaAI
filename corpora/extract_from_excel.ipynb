```python
import pandas as pd
```

```python
file = "../corpora/raw/modified_file_updated_1.xlsx"
data = pd.read_excel(file, sheet_name=None)
```

```python
data['command_1'].keys()
```

```python
for command in data.keys():
    tactic_name = data[command].keys()[0]
    data[command] = data[command].rename(columns={tactic_name:'Index', 'Unnamed: 1':'sentence', 'Unnamed: 2':tactic_name})
    data[command] = data[command].drop(columns='Index')
```

```python
data['command_1']
```

```python
df_list = list(data.values())
data_merged = pd.concat(df_list, ignore_index=True)
```

```python
data_merged.fillna(0)
```

```python
data_merged.to_csv('../corpora/all_generated_tactics.csv')
```
